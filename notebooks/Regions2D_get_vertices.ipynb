```python
import echoregions as er
```

Read EVR file.

```python
EVR_FILE = 'x1.evr'
r2d = er.read_evr(EVR_FILE)
```

For each region_id in the dataset, we print out each vertex (ping_time, depth) values.

```python
# Get all vertices from each region_id
region_id = [item for item in r2d.data["region_id"]]
ping_times = [list(item)+[item[0]] for item in r2d.data["time"]]
depths = [list(item)+[item[0]] for item in r2d.data["depth"]]
for value in list(zip(region_id, ping_times, depths)):
    print("---------------")
    print("Region ID")
    print(value[0])
    for index in range(len(value[1])):
        if index == len(value[1]) - 1:
            print(f"Closing/Repeated Vertex #{index + 1}:")
        else:
            print(f"Vertex #{index + 1}:")
        print(f"Ping Time - {value[1][index]}")
        print(f"Depth - {value[1][index]}")
```

We do the same as prior, but we select one region. We expect this region to be unique to just one region_id, and as such will only contain vertices from that one region_id.

```python
# Select region
df = r2d.select_region(region=11)
region_id = [item for item in df["region_id"]]
ping_times = [list(item)+[item[0]] for item in df["time"]]
depths = [list(item)+[item[0]] for item in df["depth"]]
for value in list(zip(region_id, ping_times, depths)):
    print("---------------")
    print("Region ID")
    print(value[0])
    for index in range(len(value[1])):
        if index == len(value[1]) - 1:
            print(f"Closing/Repeated Vertex #{index + 1}:")
        else:
            print(f"Vertex #{index + 1}:")
        print(f"Ping Time - {value[1][index]}")
        print(f"Depth - {value[1][index]}")
```
